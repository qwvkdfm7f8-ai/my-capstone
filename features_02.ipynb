#TASK 1 — One-Hot Encoding
I applied one-hot encoding to categorical columns to convert them into numerical format so they can be used in analysis and modeling.

```python
import pandas as pd
def one_hot_encode(df):
    df_encoded = pd.get_dummies(
        df,
        columns=["MS Zoning", "Neighborhood"],
        prefix=["zone", "neigh"] )

    print(df_encoded.head())
    return df_encoded
df = one_hot_encode(df)
```

#TASK 2 — Ordinal Encoding
I encoded the "Exter Qual" column using ordinal encoding because it has a natural order from poor to excellent.

```python
import pandas as pd

def Exter_Qual_Ord(df):
    exter_qual = pd.CategoricalDtype(
        categories=["Po", "Fa", "TA", "Gd", "Ex"],
        ordered=True
    )
    df["Exter_Qual_Ord"] = df["Exter Qual"].astype(exter_qual).cat.codes
    print(df[["Exter Qual", "Exter_Qual_Ord"]].head())
    return df

df=Exter_Qual_Ord(df)
```

#TASK 3 —Scale
I scaled numerical features using StandardScaler to ensure that all values are on a similar scale, improving model performance. The scaler was fit on the training data and applied to the test data.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def scale_data(df, cols_to_scale, test_size=0.2, random_state=42):
    X_train, X_test = train_test_split(df, test_size=test_size, random_state=random_state)
    scaler = StandardScaler()
    X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
    X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])

    return X_train, X_test
cols = ["Gr Liv Area", "Lot Area"]
X_train, X_test = scale_data(df, cols)
X_train.head()
```

#TASK *4* —Create Domain Features
I created new features such as price per square foot and price per bedroom to better capture the relationship between price and property size.

```python
def create_domain_features(df):
    df["price_per_sqft"] = df["SalePrice"] / (df["Gr Liv Area"] + 1e-3)
    df["price_per_bedroom"] = df["SalePrice"] / (df["Bedroom AbvGr"] + 1e-3)
    return df
df = create_domain_features(df)
print(df[["price_per_sqft", "price_per_bedroom"]].head())
```

#TASK 5 —Create Interaction Feature
I created an interaction feature by combining overall quality and living area to capture their combined effect on house price.

```python
def add_interaction_feature(df):
    df["quality_area"] = df["Overall Qual"] * df["Gr Liv Area"]
    return df
df = add_interaction_feature(df)
print(df["quality_area"].head())
```

#TASK 6 — Log-transform
I applied a log transformation to the "SalePrice" column to reduce skewness and make the distribution more normal.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def log_transform(df, col):
    df[col + "_log"] = np.log1p(df[col])
    fig, axes = plt.subplots(1, 2, figsize=(12,5))
    sns.histplot(df[col], ax=axes[0], color='skyblue', bins=30).set(title=f"{col} raw")
    sns.histplot(df[col + "_log"], ax=axes[1], color='salmon', bins=30).set(title=f"{col} _log")
    plt.show()
    return df
df = log_transform(df, "SalePrice")
```

#TASK 7 —Bin
I grouped the "Year Built" column into categories to simplify analysis and better understand different property age groups.


```python
def create_binned_feature(df, column, labels):
    df[column + "_group"] = pd.qcut(df[column], q=len(labels), labels=labels)
    return df
df = create_binned_feature(df, "Year Built", ["Old", "Medium", "New"])
print(df["Year Built_group"].value_counts())
```

#TASK 8 —Remove redundant features
I removed highly correlated features to reduce redundancy and avoid multicollinearity in the dataset.

```python
def drop_correlated_features(df, threshold=0.95):
    corr = df.corr(numeric_only=True).abs()
    upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    to_drop = [col for col in upper.columns
               if any(upper[col] > threshold) and col != "SalePrice"]
    return df.drop(columns=to_drop), to_drop
df_reduced, dropped_columns = drop_correlated_features(df)
print("Dropped:", dropped_columns)
print("Shape:", df_reduced.shape)
```

#Final Step: Create features() Function

```python
def run_feature_engineering():
    df = clean_data()
    df = one_hot_encode(df)
    df = Exter_Qual_Ord(df)
    df = create_domain_features(df)
    df = add_interaction_feature(df)
    df = log_transform(df, "SalePrice")
    df = create_binned_feature(df, "Year Built", ["Old", "Medium", "New"])
    df, dropped_columns = drop_correlated_features(df)

    cols = ["Gr Liv Area", "Lot Area"]
    X_train, X_test = scale_data(df, cols)

    return X_train, X_test, df, dropped_columns
X_train, X_test, df, dropped_columns = run_feature_engineering()
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Dropped columns:", dropped_columns)
print(df.head())
```
